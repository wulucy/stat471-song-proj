```python
def get_genres(track_id):
    
    # Get song data
    response = requests.get("https://api.spotify.com/v1/tracks/{}".format(track_id), headers=headers)
    
    if response.status_code == 200:
        retry_after = 0
        response_parsed = json.loads(response.text)
        response_artists = response_parsed['artists']

        # Get genre
        genres = list()
        for artist in response_artists:
            artist_id = artist['id']
            response = requests.get("https://api.spotify.com/v1/artists/{}".format(artist_id), headers=headers)
            response_parsed = json.loads(response.text)
            try:
                artist_genres = response_parsed['genres']
                genres += artist_genres
            except:
                print('Error response:', response_parsed)
                genres += ["000"]
    
    elif response.status_code == 429:
        print('Retry headers:', response.headers)
        retry_after = int(response.headers['Retry-After'])
        genres = ['NA']
    
    else:
        retry_after = 0
        print("Failed for track id {}".format(track_id))
        genres = ['NA']
        
    genres = ", ".join(genres)
        
    return retry_after, genres
```
